```python
df = pd.read_csv('car_price_dataset.csv')
df = df.drop(columns = ['Model', 'Brand'])
```

```python
X = df.select_dtypes(np.number)
df[X.columns] = StandardScaler().fit_transform(X)
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df, columns = ['Transmission'] )
df = pd.get_dummies(df, columns = ['Fuel_Type'])
df
```

```python
df['Transmission_Automatic'].replace({True : 1, False : 0}, inplace = True )
df['Transmission_Manual'].replace({True : 1, False : 0}, inplace = True )
df['Fuel_Type_Diesel'].replace({True : 1, False : 0}, inplace = True )
df['Fuel_Type_Electric'].replace({True : 1, False : 0}, inplace = True )
df['Fuel_Type_Petrol'].replace({True : 1, False : 0}, inplace = True )

```

```python
df =df.drop(columns = 'Transmission_Semi-Automatic')
```

```python
df = df.drop(columns = ['Fuel_Type_Hybrid' ])
```

```python
y = df['Price']
```

```python
x = df.drop(columns = 'Price')
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model
```

```python
df
```

```python
mse_train = mean_absolute_error(y_train, model.predict(x_train))
print(f" Training-set error: {mse_train}")
```

```python
mse_test = mean_absolute_error(y_test, model.predict(x_test))
print(f" Testing-set error: {mse_test}")
```
